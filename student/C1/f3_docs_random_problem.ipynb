````python
def generate_password(length: int, min_digits: int, min_symbols: int) -> str:
    lowercase = "abcdefghijklmnopqrstuvwxyz"
    uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    digits = "0123456789"
    symbols = "!@#$%^&*"

    password = []

    for _ in range(min_digits):
        password.append(random.choice(digits))

    for _ in range(min_symbols):
        password.append(random.choice(symbols))

    all_chars = lowercase + uppercase + digits + symbols

    remaining = length - len(password)
    for _ in range(remaining):
        password.append(random.choice(all_chars))

    random.shuffle(password)

    return "".join(password)
````

````python
print(generate_password(12, 2, 2))
````

````python
def generate_passphrase(words: list[str],num_words: int,separator: str,capitalize_first: bool,include_number: bool,) -> str:
    result = random.choices(words, k=num_words)

    if capitalize_first:
        result = [word.capitalize() for word in result]

    if include_number:
        result.append(str(random.randint(0, 9)))

    return separator.join(result)
````

````python
words = ["apple", "moon", "book", "cloud"]

print(generate_passphrase(words, 2, "-", True, True))
````
