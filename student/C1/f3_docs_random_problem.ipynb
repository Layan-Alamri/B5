````python
def generate_password(length, min_digits=2, min_symbols=2):
    digits = "0123456789"
    symbols = "!@#$%^&*"
    letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    result = []

    for _ in range(min_digits):
        result.append(random.choice(digits))

    for _ in range(min_symbols):
        result.append(random.choice(symbols))

    for _ in range(length - min_digits - min_symbols):
        result.append(random.choice(letters + digits + symbols))

    random.shuffle(result)
    return "".join(result)
````
